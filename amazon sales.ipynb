```python
import numpy as np
Q1 = amazon['discounted_price'].quantile(0.25)
Q3 = amazon['discounted_price'].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1-1.5*IQR
Upper_bound = Q3+1.5*IQR

print(f"Lower_bound : {lower_bound}")
print(f"upper_bound : {Upper_bound}")

#calculate the median

Median_value = amazon["discounted_price"].median()
print(f"Median Value:{Median_value}")

#replace outliers with median


Median_value = amazon["discounted_price"].median()
amazon["discounted_price"] =np.where((amazon["discounted_price"] < lower_bound)|(amazon["discounted_price"] > Upper_bound),Median_value,amazon["discounted_price"])
print("Data after Replacing Outliers with Median:\n",amazon)

```

```python
import numpy as np
Q1 = amazon['actual_price'].quantile(0.25)
Q3 = amazon['actual_price'].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1-1.5*IQR
Upper_bound = Q3+1.5*IQR

print(f"Lower_bound : {lower_bound}")
print(f"upper_bound : {Upper_bound}")

#calculate the median

Median_value = amazon["actual_price"].median()
print(f"Median Value:{Median_value}")

#replace outliers with median


Median_value = amazon["actual_price"].median()
amazon["actual_price"] =np.where((amazon["actual_price"] < lower_bound)|(amazon["actual_price"] > Upper_bound),Median_value,amazon["actual_price"])
print("Data after Replacing Outliers with Median:\n",amazon)
```

## Paired ttest

```python
import scipy.stats as stats

t_stat, p_value = stats.ttest_rel(amazon['actual_price'], amazon['discounted_price'])
alpha = 0.05
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
if p_value < alpha:
    print(f"Reject the null hypothesis at alpha ={alpha}")
else:
    print(f"Fail to reject the null hypothesis at alpha = {alpha}")
    
```
